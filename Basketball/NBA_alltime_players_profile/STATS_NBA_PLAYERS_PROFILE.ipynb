# From file - Alltimenbaplayers

```python
# Read Data from file
with open('./data/Alltimenbaplayerslist.json') as json_data:
    k = json.load(json_data)
```

```python
players_alltime = k['apiResults'][0]['league']['players']
len(players_alltime)
```

```python
for i in range(0, 3960):
    players_alltime[i]['positions'] = {'name':players_alltime[i]['positions'][0]['abbreviation']}
```

```python
from pandas.io.json import json_normalize
json_normalize(players_alltime)
```

# Output csv & clean data

```python
output = json_normalize(players_alltime)
```

```python
output.to_csv('nbaplayer_alltime.csv',index=False)
```
