### Use Python to query Bigquery

```python
import pandas
query = """
SELECT 
    player_id, pos, firstname, lastname, birth_city, birth_state, birth_country, college_abb, exp_first_year, 
    draft_round, draft_overall_pick

FROM 
    [nba.alltime_players_profile]

WHERE 
    college_abb = "North Carolina"

ORDER BY exp_first_year DESC
"""
pandas.io.gbq.read_gbq(query, project_id='sports-1211')
```
